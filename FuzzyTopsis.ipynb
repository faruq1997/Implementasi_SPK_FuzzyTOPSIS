```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder

# Inisialisasi LabelEncoder
label_encoder = LabelEncoder()

# Fungsi FuzzyTOPSIS
def fuzzy_topsis(df, criteria_weights):
    # Label Encoding untuk fitur tertentu
    df['Fasilitas'] = label_encoder.fit_transform(df['Fasilitas'])
    df['Keamanan'] = label_encoder.fit_transform(df['Keamanan'])
    df['Kebersihan'] = label_encoder.fit_transform(df['Kebersihan'])
    df['Akses Jalan'] = label_encoder.fit_transform(df['Akses Jalan'])

    # Normalisasi bobot agar total menjadi 100
    total_weight = sum(criteria_weights)
    criteria_weights = [weight / total_weight for weight in criteria_weights]


    # Step 1: Normalisasi Data
    normalized_df = df.copy()
    for column in df.columns[1:]:
        normalized_df[column] = df[column] / np.linalg.norm(df[column])

    # Step 2: Fuzzyfikasi Data
    fuzzy_df = normalized_df.copy()
    for column in normalized_df.columns[1:]:
        fuzzy_df[column] = np.exp(-1 / (normalized_df[column] ** 2))

    # Step 3: Penentuan Matriks Keputusan
    decision_matrix = fuzzy_df.iloc[:, 1:].values

    # Step 4: Penentuan Matriks Bobot
    weight_matrix = np.array(criteria_weights)

    # Step 5: Perhitungan Solusi Ideal Positif dan Negatif
    ideal_positive = np.max(decision_matrix, axis=0)
    ideal_negative = np.min(decision_matrix, axis=0)

    # Step 6: Perhitungan Jarak Alternatif terhadap Solusi Ideal Positif dan Negatif
    positive_distance = np.linalg.norm(decision_matrix - ideal_positive, axis=1)
    negative_distance = np.linalg.norm(decision_matrix - ideal_negative, axis=1)

    # Step 7: Perangkingan Alternatif
    closeness_coefficient = negative_distance / (negative_distance + positive_distance)
    df['Closeness Coefficient'] = closeness_coefficient
    ranked_df = df.sort_values(by='Closeness Coefficient', ascending=False).reset_index(drop=True)

    return ranked_df

criteria_weights = [30, 15, 20, 15, 20]
# Jalankan FuzzyTOPSIS pada DataFrame df_wisata
result_df = fuzzy_topsis(df_wisata, criteria_weights)

# Tampilkan hasil ranking
print(result_df[['Nama', 'Closeness Coefficient']])
```
